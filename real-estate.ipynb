```python

plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

print("üèõÔ∏è ADVANCED GOVERNMENT REAL ESTATE FINANCIAL ANALYSIS")
print("=" * 60)
print("üìä Comprehensive Data Science Project")
print("üéØ Investment Optimization & Policy Decision Support")
print("=" * 60)

# ================================================================
# 1. DATA GENERATION & SIMULATION
# ================================================================

def generate_government_realestate_data(n_samples=5000):
    """
    Generate realistic government real estate project data
    with complex financial relationships and market dynamics
    """
    np.random.seed(42)
    
    # Base features
    project_types = ['Housing', 'Commercial', 'Infrastructure', 'Mixed-Use', 'Educational', 'Healthcare']
    regions = ['Urban', 'Suburban', 'Rural', 'Metropolitan', 'Coastal']
    funding_sources = ['Federal', 'State', 'Local', 'PPP', 'Mixed']
    
    data = {
        'project_id': range(1, n_samples + 1),
        'project_type': np.random.choice(project_types, n_samples),
        'region': np.random.choice(regions, n_samples),
        'funding_source': np.random.choice(funding_sources, n_samples),
        
        # Financial metrics (in millions USD)
        'initial_budget': np.random.lognormal(mean=2.5, sigma=0.8, size=n_samples),
        'land_cost': np.random.lognormal(mean=1.8, sigma=0.6, size=n_samples),
        'construction_cost': np.random.lognormal(mean=2.2, sigma=0.7, size=n_samples),
        'administrative_cost': np.random.lognormal(mean=0.5, sigma=0.4, size=n_samples),
        
        # Project characteristics
        'project_duration_months': np.random.normal(24, 8, n_samples).clip(6, 120),
        'total_area_sqft': np.random.lognormal(mean=10, sigma=1.2, size=n_samples),
        'population_served': np.random.lognormal(mean=8, sigma=1.5, size=n_samples),
        
        # Economic indicators
        'local_gdp_growth': np.random.normal(2.5, 1.2, n_samples),
        'unemployment_rate': np.random.normal(6.5, 2.0, n_samples).clip(1, 15),
        'median_income': np.random.lognormal(mean=10.8, sigma=0.3, size=n_samples),
        'property_tax_rate': np.random.normal(1.2, 0.3, n_samples).clip(0.5, 2.5),
        
        # Market conditions
        'interest_rate': np.random.normal(4.5, 1.5, n_samples).clip(1, 10),
        'inflation_rate': np.random.normal(2.8, 1.0, n_samples).clip(0, 8),
        'construction_price_index': np.random.normal(105, 15, n_samples),
        
        # Performance metrics
        'completion_delay_months': np.random.exponential(2, n_samples),
        'quality_score': np.random.beta(2, 1, n_samples) * 100,
        'community_satisfaction': np.random.beta(3, 2, n_samples) * 100,
    }
    
    df = pd.DataFrame(data)
    
    # Create complex target variable: Total Project Cost with realistic relationships
    cost_multipliers = {
        'Housing': 1.0, 'Commercial': 1.3, 'Infrastructure': 1.8,
        'Mixed-Use': 1.4, 'Educational': 1.1, 'Healthcare': 1.6
    }
    
    region_multipliers = {
        'Urban': 1.5, 'Suburban': 1.0, 'Rural': 0.7,
        'Metropolitan': 1.8, 'Coastal': 2.0
    }
    
    # Complex cost calculation with interactions
    df['base_cost'] = (df['initial_budget'] + 
                      df['land_cost'] * df['region'].map(region_multipliers) +
                      df['construction_cost'] * df['project_type'].map(cost_multipliers) +
                      df['administrative_cost'])
    
    # Add market condition impacts
    df['market_adjustment'] = (1 + df['inflation_rate']/100) * (1 + df['interest_rate']/100 * 0.1)
    
    # Add complexity factors
    df['complexity_factor'] = (df['total_area_sqft'] / 50000) * (df['project_duration_months'] / 24)
    
    # Final cost calculation with noise
    df['total_project_cost'] = (df['base_cost'] * df['market_adjustment'] * 
                               (1 + df['complexity_factor'] * 0.2) * 
                               (1 + df['completion_delay_months'] * 0.05) + 
                               np.random.normal(0, df['base_cost'] * 0.1))
    
    # Calculate derived financial metrics
    df['cost_per_sqft'] = df['total_project_cost'] * 1000000 / df['total_area_sqft']
    df['cost_overrun_pct'] = ((df['total_project_cost'] - df['initial_budget']) / df['initial_budget'] * 100)
    df['roi_estimate'] = np.random.normal(5, 3, n_samples) + df['quality_score'] * 0.1
    df['npv_estimate'] = df['total_project_cost'] * (df['roi_estimate'] / 100) * 10
    
    # Add temporal dimension
    start_dates = pd.date_range('2015-01-01', '2023-12-31', periods=n_samples)
    df['project_start_date'] = start_dates
    df['project_end_date'] = df['project_start_date'] + pd.to_timedelta(df['project_duration_months'] * 30, unit='days')
    df['year'] = df['project_start_date'].dt.year
    df['quarter'] = df['project_start_date'].dt.quarter
    
    # Clean up intermediate columns
    df = df.drop(['base_cost', 'market_adjustment', 'complexity_factor'], axis=1)
    
    return df

# Generate the dataset
print("üîÑ Generating synthetic government real estate dataset...")
df = generate_government_realestate_data(5000)
print(f"‚úÖ Dataset created: {df.shape[0]} projects with {df.shape[1]} features\n")

# ================================================================
# 2. ADVANCED EXPLORATORY DATA ANALYSIS (EDA)
# ================================================================

print("üìä ADVANCED EXPLORATORY DATA ANALYSIS")
print("=" * 40)

# Dataset overview
print("üìã Dataset Overview:")
print(f"Shape: {df.shape}")
print(f"Memory usage: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
print(f"Date range: {df['project_start_date'].min()} to {df['project_start_date'].max()}")
print()

# Statistical summary
print("üìà Statistical Summary:")
numeric_cols = df.select_dtypes(include=[np.number]).columns
print(df[numeric_cols].describe().round(2))
print()

# Missing values analysis
print("üîç Data Quality Assessment:")
missing_data = pd.DataFrame({
    'Column': df.columns,
    'Missing_Count': df.isnull().sum(),
    'Missing_Percentage': (df.isnull().sum() / len(df)) * 100,
    'Data_Type': df.dtypes
})
print(missing_data[missing_data['Missing_Count'] > 0])
print()

# Advanced visualizations
fig, axes = plt.subplots(2, 2, figsize=(20, 16))

# 1. Distribution of total project costs
axes[0, 0].hist(df['total_project_cost'], bins=50, alpha=0.7, color='skyblue', edgecolor='black')
axes[0, 0].axvline(df['total_project_cost'].mean(), color='red', linestyle='--', 
                   label=f'Mean: ${df["total_project_cost"].mean():.2f}M')
axes[0, 0].axvline(df['total_project_cost'].median(), color='green', linestyle='--',
                   label=f'Median: ${df["total_project_cost"].median():.2f}M')
axes[0, 0].set_title('Distribution of Total Project Costs', fontsize=14, fontweight='bold')
axes[0, 0].set_xlabel('Total Project Cost (Millions USD)')
axes[0, 0].set_ylabel('Frequency')
axes[0, 0].legend()

# 2. Cost by project type
cost_by_type = df.groupby('project_type')['total_project_cost'].agg(['mean', 'median', 'std']).reset_index()
x_pos = np.arange(len(cost_by_type))
axes[0, 1].bar(x_pos, cost_by_type['mean'], yerr=cost_by_type['std'], 
               capsize=5, alpha=0.8, color='lightcoral')
axes[0, 1].set_xticks(x_pos)
axes[0, 1].set_xticklabels(cost_by_type['project_type'], rotation=45, ha='right')
axes[0, 1].set_title('Average Project Cost by Type', fontsize=14, fontweight='bold')
axes[0, 1].set_ylabel('Average Cost (Millions USD)')

# 3. Cost overrun analysis
axes[1, 0].scatter(df['initial_budget'], df['cost_overrun_pct'], alpha=0.6, color='orange')
axes[1, 0].axhline(y=0, color='red', linestyle='--', alpha=0.8)
axes[1, 0].set_title('Cost Overruns vs Initial Budget', fontsize=14, fontweight='bold')
axes[1, 0].set_xlabel('Initial Budget (Millions USD)')
axes[1, 0].set_ylabel('Cost Overrun (%)')

# 4. Time series of project costs
yearly_costs = df.groupby('year')['total_project_cost'].agg(['mean', 'count']).reset_index()
ax2 = axes[1, 1].twinx()
line1 = axes[1, 1].plot(yearly_costs['year'], yearly_costs['mean'], marker='o', 
                        color='blue', linewidth=2, label='Avg Cost')
bar1 = ax2.bar(yearly_costs['year'], yearly_costs['count'], alpha=0.3, 
               color='gray', label='Project Count')
axes[1, 1].set_title('Project Costs and Volume Over Time', fontsize=14, fontweight='bold')
axes[1, 1].set_xlabel('Year')
axes[1, 1].set_ylabel('Average Cost (Millions USD)', color='blue')
ax2.set_ylabel('Number of Projects', color='gray')
axes[1, 1].tick_params(axis='y', labelcolor='blue')
ax2.tick_params(axis='y', labelcolor='gray')

plt.tight_layout()
plt.show()

# Advanced correlation analysis
print("üîó CORRELATION ANALYSIS")
print("=" * 25)

# Select key numeric features for correlation
key_features = ['total_project_cost', 'initial_budget', 'land_cost', 'construction_cost',
                'project_duration_months', 'total_area_sqft', 'local_gdp_growth',
                'unemployment_rate', 'median_income', 'interest_rate', 'inflation_rate',
                'completion_delay_months', 'quality_score', 'cost_overrun_pct', 'roi_estimate']

correlation_matrix = df[key_features].corr()

# Create advanced correlation heatmap
plt.figure(figsize=(16, 12))
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
sns.heatmap(correlation_matrix, mask=mask, annot=True, cmap='RdYlBu_r', center=0,
            square=True, linewidths=0.5, cbar_kws={"shrink": 0.8}, fmt='.2f')
plt.title('Advanced Correlation Matrix - Government Real Estate Projects', 
          fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()

# Identify strong correlations
strong_corr_pairs = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        corr_val = correlation_matrix.iloc[i, j]
        if abs(corr_val) > 0.5:  # Strong correlation threshold
            strong_corr_pairs.append({
                'Feature_1': correlation_matrix.columns[i],
                'Feature_2': correlation_matrix.columns[j],
                'Correlation': corr_val
            })

strong_corr_df = pd.DataFrame(strong_corr_pairs).sort_values('Correlation', key=abs, ascending=False)
print("üí™ Strong Correlations (|r| > 0.5):")
print(strong_corr_df.head(10))
print()

# ================================================================
# 3. ADVANCED FEATURE ENGINEERING
# ================================================================

print("‚öôÔ∏è ADVANCED FEATURE ENGINEERING")
print("=" * 35)

def engineer_features(df):
    """Advanced feature engineering for government real estate projects"""
    df_eng = df.copy()
    
    # 1. Financial ratios and efficiency metrics
    df_eng['cost_efficiency'] = df_eng['total_project_cost'] / df_eng['total_area_sqft']
    df_eng['budget_utilization'] = df_eng['total_project_cost'] / df_eng['initial_budget']
    df_eng['land_cost_ratio'] = df_eng['land_cost'] / df_eng['total_project_cost']
    df_eng['construction_cost_ratio'] = df_eng['construction_cost'] / df_eng['total_project_cost']
    df_eng['admin_cost_ratio'] = df_eng['administrative_cost'] / df_eng['total_project_cost']
    
    # 2. Economic impact features
    df_eng['cost_per_person_served'] = df_eng['total_project_cost'] * 1000000 / df_eng['population_served']
    df_eng['economic_value_index'] = (df_eng['local_gdp_growth'] * df_eng['median_income']) / df_eng['unemployment_rate']
    df_eng['market_favorability'] = (10 - df_eng['interest_rate']) * (5 - df_eng['inflation_rate'])
    
    # 3. Project complexity and risk features
    df_eng['project_complexity_score'] = (
        (df_eng['total_area_sqft'] / df_eng['total_area_sqft'].median()) * 0.3 +
        (df_eng['project_duration_months'] / df_eng['project_duration_months'].median()) * 0.3 +
        (df_eng['population_served'] / df_eng['population_served'].median()) * 0.4
    )
    
    df_eng['risk_score'] = (
        df_eng['completion_delay_months'] * 0.4 +
        (100 - df_eng['quality_score']) * 0.3 +
        abs(df_eng['cost_overrun_pct']) * 0.3
    )
    
    # 4. Temporal features
    df_eng['project_year'] = df_eng['project_start_date'].dt.year
    df_eng['project_month'] = df_eng['project_start_date'].dt.month
    df_eng['project_season'] = df_eng['project_month'].map({
        12: 'Winter', 1: 'Winter', 2: 'Winter',
        3: 'Spring', 4: 'Spring', 5: 'Spring',
        6: 'Summer', 7: 'Summer', 8: 'Summer',
        9: 'Fall', 10: 'Fall', 11: 'Fall'
    })
    
    # 5. Performance benchmarking
    df_eng['cost_vs_regional_avg'] = df_eng.groupby('region')['total_project_cost'].transform(
        lambda x: (x - x.mean()) / x.std()
    )
    df_eng['cost_vs_type_avg'] = df_eng.groupby('project_type')['total_project_cost'].transform(
        lambda x: (x - x.mean()) / x.std()
    )
    
    # 6. Interaction features
    df_eng['region_type_interaction'] = df_eng['region'] + '_' + df_eng['project_type']
    df_eng['funding_region_interaction'] = df_eng['funding_source'] + '_' + df_eng['region']
    
    # 7. Polynomial features for non-linear relationships
    df_eng['area_squared'] = df_eng['total_area_sqft'] ** 2
    df_eng['duration_squared'] = df_eng['project_duration_months'] ** 2
    df_eng['budget_duration_interaction'] = df_eng['initial_budget'] * df_eng['project_duration_months']
    
    print(f"‚úÖ Feature engineering completed. New shape: {df_eng.shape}")
    print(f"üìà Added {df_eng.shape[1] - df.shape[1]} new features")
    
    return df_eng

# Apply feature engineering
df_engineered = engineer_features(df)

# Display new features
new_features = [col for col in df_engineered.columns if col not in df.columns]
print(f"\nüÜï New Features Created ({len(new_features)}):")
for i, feat in enumerate(new_features, 1):
    print(f"{i:2d}. {feat}")

# ================================================================
# 4. ADVANCED CLUSTERING ANALYSIS
# ================================================================

print("\nüéØ ADVANCED CLUSTERING ANALYSIS")
print("=" * 35)

# Prepare data for clustering
cluster_features = ['total_project_cost', 'cost_efficiency', 'project_complexity_score',
                   'risk_score', 'quality_score', 'roi_estimate', 'economic_value_index']

# Standardize features
scaler = StandardScaler()
X_cluster = scaler.fit_transform(df_engineered[cluster_features])

# Determine optimal number of clusters using elbow method
inertias = []
silhouette_scores = []
k_range = range(2, 11)

from sklearn.metrics import silhouette_score

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_cluster)
    inertias.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_cluster, kmeans.labels_))

# Plot elbow curve and silhouette scores
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

ax1.plot(k_range, inertias, marker='o', linewidth=2, markersize=8)
ax1.set_title('Elbow Method for Optimal K', fontsize=14, fontweight='bold')
ax1.set_xlabel('Number of Clusters (k)')
ax1.set_ylabel('Inertia')
ax1.grid(True, alpha=0.3)

ax2.plot(k_range, silhouette_scores, marker='s', linewidth=2, markersize=8, color='red')
ax2.set_title('Silhouette Score vs Number of Clusters', fontsize=14, fontweight='bold')
ax2.set_xlabel('Number of Clusters (k)')
ax2.set_ylabel('Silhouette Score')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Perform clustering with optimal k
optimal_k = k_range[np.argmax(silhouette_scores)]
print(f"üéØ Optimal number of clusters: {optimal_k}")

kmeans_final = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
df_engineered['cluster'] = kmeans_final.fit_predict(X_cluster)

# Analyze clusters
print(f"\nüìä Cluster Analysis (k={optimal_k}):")
cluster_summary = df_engineered.groupby('cluster')[cluster_features].agg(['mean', 'std']).round(2)
print(cluster_summary)

# Cluster distribution
cluster_dist = df_engineered['cluster'].value_counts().sort_index()
print(f"\nüìà Cluster Distribution:")
for cluster, count in cluster_dist.items():
    pct = count / len(df_engineered) * 100
    print(f"Cluster {cluster}: {count} projects ({pct:.1f}%)")

# Visualize clusters using PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_cluster)

plt.figure(figsize=(12, 8))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=df_engineered['cluster'], 
                     cmap='viridis', alpha=0.6, s=50)
plt.colorbar(scatter)
plt.title('Project Clusters Visualization (PCA)', fontsize=16, fontweight='bold')
plt.xlabel(f'First Principal Component (Explained Variance: {pca.explained_variance_ratio_[0]:.2%})')
plt.ylabel(f'Second Principal Component (Explained Variance: {pca.explained_variance_ratio_[1]:.2%})')

# Add cluster centers
centers_pca = pca.transform(scaler.transform(
    df_engineered.groupby('cluster')[cluster_features].mean()
))
plt.scatter(centers_pca[:, 0], centers_pca[:, 1], c='red', marker='x', s=200, linewidths=3)
plt.grid(True, alpha=0.3)
plt.show()

# ================================================================
# 5. ADVANCED PREDICTIVE MODELING
# ================================================================

print("ü§ñ ADVANCED PREDICTIVE MODELING")
print("=" * 35)

# Prepare features for modeling
# Encode categorical variables
le_dict = {}
categorical_cols = ['project_type', 'region', 'funding_source', 'project_season', 
                   'region_type_interaction', 'funding_region_interaction']

df_model = df_engineered.copy()
for col in categorical_cols:
    le = LabelEncoder()
    df_model[col + '_encoded'] = le.fit_transform(df_model[col])
    le_dict[col] = le

# Select features for modeling
feature_cols = [
    # Original features
    'initial_budget', 'land_cost', 'construction_cost', 'administrative_cost',
    'project_duration_months', 'total_area_sqft', 'population_served',
    'local_gdp_growth', 'unemployment_rate', 'median_income', 'property_tax_rate',
    'interest_rate', 'inflation_rate', 'construction_price_index',
    'completion_delay_months', 'quality_score', 'community_satisfaction',
    
    # Engineered features
    'cost_efficiency', 'budget_utilization', 'land_cost_ratio', 'construction_cost_ratio',
    'admin_cost_ratio', 'cost_per_person_served', 'economic_value_index',
    'market_favorability', 'project_complexity_score', 'risk_score',
    'cost_vs_regional_avg', 'cost_vs_type_avg', 'area_squared', 'duration_squared',
    'budget_duration_interaction', 'project_year', 'project_month', 'cluster',
    
    # Encoded categorical features
    'project_type_encoded', 'region_encoded', 'funding_source_encoded',
    'project_season_encoded', 'region_type_interaction_encoded', 'funding_region_interaction_encoded'
]

X = df_model[feature_cols]
y = df_model['total_project_cost']

print(f"üìä Modeling Dataset:")
print(f"Features: {X.shape[1]}")
print(f"Samples: {X.shape[0]}")
print(f"Target: {y.name}")

# Check for multicollinearity using VIF
def calculate_vif(df, features):
    """Calculate Variance Inflation Factor for features"""
    vif_data = pd.DataFrame()
    vif_data["Feature"] = features
    vif_data["VIF"] = [variance_inflation_factor(df.values, i) for i in range(len(features))]
    return vif_data.sort_values('VIF', ascending=False)

# Select subset of features for VIF calculation (to avoid computational issues)
vif_features = ['initial_budget', 'construction_cost', 'project_duration_months', 
               'total_area_sqft', 'interest_rate', 'inflation_rate', 'quality_score',
               'project_complexity_score', 'risk_score', 'economic_value_index']

vif_scores = calculate_vif(X[vif_features], vif_features)
print(f"\nüìà Multicollinearity Analysis (VIF > 5 indicates high correlation):")
print(vif_scores.head(10))

# Feature selection using statistical methods
selector = SelectKBest(score_func=f_regression, k=30)
X_selected = selector.fit_transform(X, y)
selected_features = [feature_cols[i] for i in selector.get_support(indices=True)]

print(f"\nüéØ Top 30 Features Selected:")
feature_scores = pd.DataFrame({
    'Feature': selected_features,
    'Score': selector.scores_[selector.get_support()]
}).sort_values('Score', ascending=False)
print(feature_scores.head(15))

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X_selected, y, test_size=0.2, random_state=42, stratify=df_model['cluster']
)

# Scale features
scaler = RobustScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print(f"\nüìä Data Split:")
print(f"Training set: {X_train.shape[0]} samples")
print(f"Test set: {X_test.shape[0]} samples")

# ================================================================
# 6. MODEL TRAINING AND EVALUATION
# ================================================================

print("\nüèãÔ∏è MODEL TRAINING AND EVALUATION")
print("=" * 40)

# Define models
models = {
    'Ridge Regression': Ridge(alpha=1.0),
    'Lasso Regression': Lasso(alpha=0.1),
    'Elastic Net': ElasticNet(alpha=0.1, l1_ratio=0.5),
    'Random Forest': RandomForestRegressor(n_estimators=100, random_state=42),
    'Gradient Boosting': GradientBoostingRegressor(n_estimators=100, random_state=42),
    'XGBoost': xgb.XGBRegressor(n_estimators=100, random_state=42),
    'LightGBM': lgb.LGBMRegressor(n_estimators=100, random_state=42, verbose=-1),
    'SVR': SVR(kernel='rbf', C=1.0)
}

# Train and evaluate models
results = {}
cv_scores = {}

print("üîÑ Training models...")
for name, model in models.items():
    print(f"Training {name}...")
    
    # Cross-validation
    cv_score = cross_val_score(model, X_train_scaled, y_train, cv=5, 
                              scoring='neg_mean_squared_error', n_jobs=-1)
    cv_scores[name] = -cv_score.mean()
    
    # Fit and predict
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)
    
    # Calculate metrics
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)
    
    results[name] = {
        'MAE': mae,
        'MSE': mse,
        'RMSE': rmse,
        'R¬≤': r2,
        'CV_RMSE': np.sqrt(cv_scores[name])
    }

# Display results
results_df = pd.DataFrame(results).T.round(4)
results_df = results_df.sort_values('R¬≤', ascending=False)

print("\nüìä MODEL PERFORMANCE COMPARISON:")
print("=" * 50)
print(results_df)

# Best model
best_model_name = results_df.index[0]
best_model = models[best_model_name]
print(f"\nüèÜ Best Model: {best_model_name}")
print(f"   R¬≤ Score: {results_df.loc[best_model_name, 'R¬≤']:.4f}")
print(f"   RMSE: {results_df.loc[best_model_name, 'RMSE']:.4f}")

# Visualize model performance
fig, axes = plt.subplots(2, 2, figsize=(18, 14))

# 1. Model comparison bar chart
model_names = results_df.index
r2_scores = results_df['R¬≤']
colors = plt.cm.viridis(np.linspace(0, 1, len(model_names)))

bars = axes[0, 0].bar(range(len(model_names)), r2_scores, color=colors, alpha=0.8)
axes[0, 0].set_xticks(range(len(model_names)))
axes[0, 0].set_xticklabels(model_names, rotation=45, ha='right')
axes[0, 0].set_title('Model Performance Comparison (R¬≤ Score)', fontsize=14, fontweight='bold')
axes[0, 0].set_ylabel('R¬≤ Score')
axes[0, 0].grid(True, alpha=0.3)

# Add value labels on bars
for bar, score in zip(bars, r2_scores):
    height = bar.get_height()
    axes[0, 0].text(bar.get_x() + bar.get_width()/2., height + 0.01,
                    f'{score:.3f}', ha='center', va='bottom', fontweight='bold')

# 2. Actual vs Predicted for best model
best_model.fit(X_train_scaled, y_train)
y_pred_best = best_model.predict(X_test_scaled)

axes[0, 1].scatter(y_test, y_pred_best, alpha=0.6, color='blue', s=30)
min_val = min(y_test.min(), y_pred_best.min())
max_val = max(y_test.max(), y_pred_best.max())
axes[0, 1].plot([min_val, max_val], [min_val, max_val], 'r--', lw=2)
axes[0, 1].set_title(f'Actual vs Predicted - {best_model_name}', fontsize=14, fontweight='bold')
axes[0, 1].set_xlabel('Actual Total Project Cost')
axes[0, 1].set_ylabel('Predicted Total Project Cost')
axes[0, 1].grid(True, alpha=0.3)

# Add R¬≤ annotation
axes[0, 1].text(0.05, 0.95, f'R¬≤ = {r2_score(y_test, y_pred_best):.3f}', 
                transform=axes[0, 1].transAxes, fontsize=12, fontweight='bold',
                bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

# 3. Residual plot
residuals = y_test - y_pred_best
axes[1, 0].scatter(y_pred_best, residuals, alpha=0.6, color='green', s=30)
axes[1, 0].axhline(y=0, color='red', linestyle='--', lw=2)
axes[1, 0].set_title('Residual Plot', fontsize=14, fontweight='bold')
axes[1, 0].set_xlabel('Predicted Values')
axes[1, 0].set_ylabel('Residuals')
axes[1, 0].grid(True, alpha=0.3)

# 4. RMSE comparison
rmse_scores = results_df['RMSE']
axes[1, 1].barh(range(len(model_names)), rmse_scores, color=colors, alpha=0.8)
axes[1, 1].set_yticks(range(len(model_names)))
axes[1, 1].set_yticklabels(model_names)
axes[1, 1].set_title('Model RMSE Comparison', fontsize=14, fontweight='bold')
axes[1, 1].set_xlabel('RMSE')
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# ================================================================
# 7. ADVANCED HYPERPARAMETER TUNING
# ================================================================

print("\nüîß ADVANCED HYPERPARAMETER TUNING")
print("=" * 40)

# Hyperparameter tuning for top 3 models
top_models = results_df.head(3).index.tolist()

tuned_results = {}

for model_name in top_models:
    print(f"\nüîÑ Tuning {model_name}...")
    
    if model_name == 'Random Forest':
        param_grid = {
            'n_estimators': [100, 200, 300],
            'max_depth': [10, 20, None],
            'min_samples_split': [2, 5, 10],
            'min_samples_leaf': [1, 2, 4]
        }
        model = RandomForestRegressor(random_state=42)
        
    elif model_name == 'XGBoost':
        param_grid = {
            'n_estimators': [100, 200, 300],
            'max_depth': [3, 6, 9],
            'learning_rate': [0.01, 0.1, 0.2],
            'subsample': [0.8, 0.9, 1.0]
        }
        model = xgb.XGBRegressor(random_state=42)
        
    elif model_name == 'Gradient Boosting':
        param_grid = {
            'n_estimators': [100, 200, 300],
            'max_depth': [3, 6, 9],
            'learning_rate': [0.01, 0.1, 0.2],
            'min_samples_split': [2, 5, 10]
        }
        model = GradientBoostingRegressor(random_state=42)
        
    elif model_name == 'LightGBM':
        param_grid = {
            'n_estimators': [100, 200, 300],
            'max_depth': [3, 6, 9],
            'learning_rate': [0.01, 0.1, 0.2],
            'num_leaves': [31, 50, 100]
        }
        model = lgb.LGBMRegressor(random_state=42, verbose=-1)
        
    elif model_name == 'Ridge Regression':
        param_grid = {
            'alpha': [0.1, 1.0, 10.0, 100.0],
            'solver': ['auto', 'saga', 'lsqr']
        }
        model = Ridge(random_state=42)
        
    elif model_name == 'Lasso Regression':
        param_grid = {
            'alpha': [0.01, 0.1, 1.0, 10.0],
            'max_iter': [1000, 2000, 5000]
        }
        model = Lasso(random_state=42)
        
    elif model_name == 'Elastic Net':
        param_grid = {
            'alpha': [0.01, 0.1, 1.0],
            'l1_ratio': [0.1, 0.5, 0.7, 0.9],
            'max_iter': [1000, 2000]
        }
        model = ElasticNet(random_state=42)
        
    elif model_name == 'SVR':
        param_grid = {
            'C': [0.1, 1.0, 10.0],
            'kernel': ['rbf', 'linear'],
            'gamma': ['scale', 'auto', 0.001, 0.01]
        }
        model = SVR()
        
    else:
        # Default parameter grid for unknown models
        param_grid = {}
        model = models[model_name]
    
    # Time series cross-validation for more robust evaluation
    tscv = TimeSeriesSplit(n_splits=5)
    
    # Grid search with time series CV
    grid_search = GridSearchCV(
        model, param_grid, cv=tscv, scoring='neg_mean_squared_error',
        n_jobs=-1, verbose=0
    )
    
    grid_search.fit(X_train_scaled, y_train)
    
    # Evaluate tuned model
    best_model_tuned = grid_search.best_estimator_
    y_pred_tuned = best_model_tuned.predict(X_test_scaled)
    
    mae_tuned = mean_absolute_error(y_test, y_pred_tuned)
    mse_tuned = mean_squared_error(y_test, y_pred_tuned)
    rmse_tuned = np.sqrt(mse_tuned)
    r2_tuned = r2_score(y_test, y_pred_tuned)
    
    tuned_results[model_name] = {
        'Best_Params': grid_search.best_params_,
        'MAE': mae_tuned,
        'RMSE': rmse_tuned,
        'R¬≤': r2_tuned,
        'CV_Score': -grid_search.best_score_
    }
    
    print(f"‚úÖ Best parameters: {grid_search.best_params_}")
    print(f"   Tuned R¬≤ Score: {r2_tuned:.4f}")
    print(f"   Improvement: {r2_tuned - results_df.loc[model_name, 'R¬≤']:.4f}")

# Display tuning results
print("\nüìä HYPERPARAMETER TUNING RESULTS:")
print("=" * 45)
tuned_df = pd.DataFrame({k: {metric: v[metric] for metric in ['MAE', 'RMSE', 'R¬≤', 'CV_Score']} 
                        for k, v in tuned_results.items()}).T.round(4)
print(tuned_df)

# ================================================================
# 8. ENSEMBLE MODELING
# ================================================================

print("\nüé≠ ENSEMBLE MODELING")
print("=" * 25)

# Create ensemble of top performing models
ensemble_models = []
ensemble_names = []

for model_name in top_models:
    if model_name in tuned_results:
        if model_name == 'Random Forest':
            model = RandomForestRegressor(**tuned_results[model_name]['Best_Params'], random_state=42)
        elif model_name == 'XGBoost':
            model = xgb.XGBRegressor(**tuned_results[model_name]['Best_Params'], random_state=42)
        elif model_name == 'Gradient Boosting':
            model = GradientBoostingRegressor(**tuned_results[model_name]['Best_Params'], random_state=42)
        elif model_name == 'Ridge Regression':
            model = Ridge(**tuned_results[model_name]['Best_Params'], random_state=42)
        
        ensemble_models.append((model_name.lower().replace(' ', '_'), model))
        ensemble_names.append(model_name)

# Voting Regressor Ensemble
voting_regressor = VotingRegressor(estimators=ensemble_models)
voting_regressor.fit(X_train_scaled, y_train)
y_pred_ensemble = voting_regressor.predict(X_test_scaled)

# Evaluate ensemble
mae_ensemble = mean_absolute_error(y_test, y_pred_ensemble)
mse_ensemble = mean_squared_error(y_test, y_pred_ensemble)
rmse_ensemble = np.sqrt(mse_ensemble)
r2_ensemble = r2_score(y_test, y_pred_ensemble)

print(f"üé≠ Ensemble Model Performance:")
print(f"   MAE: {mae_ensemble:.4f}")
print(f"   RMSE: {rmse_ensemble:.4f}")
print(f"   R¬≤ Score: {r2_ensemble:.4f}")

# Compare with individual models
ensemble_comparison = pd.DataFrame({
    'Model': ensemble_names + ['Ensemble'],
    'R¬≤': [tuned_results[name]['R¬≤'] for name in ensemble_names] + [r2_ensemble],
    'RMSE': [tuned_results[name]['RMSE'] for name in ensemble_names] + [rmse_ensemble]
})

print(f"\nüìä Ensemble vs Individual Models:")
print(ensemble_comparison)

# ================================================================
# 9. FEATURE IMPORTANCE ANALYSIS
# ================================================================

# print("\nüéØ FEATURE IMPORTANCE ANALYSIS")
# print("=" * 35)

# # Get feature importance from the best tree-based model
# if 'Random Forest' in top_models:
#     rf_model = RandomForestRegressor(**tuned_results['Random Forest']['Best_Params'], random_state=42)
#     rf_model.fit(X_train_scaled, y_train)
#     feature_importance = rf_model.feature_importances_
# elif 'XGBoost' in top_models:
#     xgb_model = xgb.XGBRegressor(**tuned_results['XGBoost']['Best_Params'], random_state=42)
#     xgb_model.fit(X_train_scaled, y_train)
#     feature_importance = xgb_model.feature_importances_

# elif 'Ridge Regression' in top_models:
#     r_reg = Ridge(**tuned_results['Ridge Regression']['Best_Params'], random_state=42)
#     r_reg.fit(X_train_scaled, y_train)
#     feature_importance = r_reg.feature_importances_
    

# # Create feature importance dataframe
# importance_df = pd.DataFrame({
#     'Feature': selected_features,
#     'Importance': feature_importance
# }).sort_values('Importance', ascending=False)

# print("üîù Top 20 Most Important Features:")
# print(importance_df.head(20))

# # Visualize feature importance
# plt.figure(figsize=(12, 10))
# top_features = importance_df.head(15)
# plt.barh(range(len(top_features)), top_features['Importance'], color='skyblue', alpha=0.8)
# plt.yticks(range(len(top_features)), top_features['Feature'])
# plt.title('Top 15 Feature Importances', fontsize=16, fontweight='bold')
# plt.xlabel('Importance')
# plt.gca().invert_yaxis()
# plt.grid(True, alpha=0.3)
# plt.tight_layout()
# plt.show()

# ================================================================
# 10. ADVANCED BUSINESS INSIGHTS & RECOMMENDATIONS
# ================================================================

print("\nüíº ADVANCED BUSINESS INSIGHTS & RECOMMENDATIONS")
print("=" * 55)

# Cost efficiency analysis by cluster
print("üí∞ COST EFFICIENCY ANALYSIS BY PROJECT CLUSTER:")
cluster_analysis = df_engineered.groupby('cluster').agg({
    'total_project_cost': ['mean', 'std', 'count'],
    'cost_efficiency': ['mean', 'std'],
    'roi_estimate': ['mean', 'std'],
    'risk_score': ['mean', 'std'],
    'quality_score': ['mean', 'std'],
    'cost_overrun_pct': ['mean', 'std']
}).round(2)

cluster_analysis.columns = ['_'.join(col).strip() for col in cluster_analysis.columns]
print(cluster_analysis)

# Regional performance analysis
print(f"\nüèóÔ∏è REGIONAL PERFORMANCE ANALYSIS:")
regional_analysis = df_engineered.groupby('region').agg({
    'total_project_cost': 'mean',
    'cost_overrun_pct': 'mean',
    'completion_delay_months': 'mean',
    'quality_score': 'mean',
    'roi_estimate': 'mean'
}).round(2).sort_values('total_project_cost', ascending=False)

print(regional_analysis)

# Project type ROI analysis
print(f"\nüìä PROJECT TYPE ROI ANALYSIS:")
type_roi_analysis = df_engineered.groupby('project_type').agg({
    'roi_estimate': 'mean',
    'total_project_cost': 'mean',
    'quality_score': 'mean',
    'risk_score': 'mean'
}).round(2).sort_values('roi_estimate', ascending=False)

print(type_roi_analysis)

# Risk factors identification
print(f"\n‚ö†Ô∏è KEY RISK FACTORS IDENTIFICATION:")
high_risk_projects = df_engineered[df_engineered['risk_score'] > df_engineered['risk_score'].quantile(0.8)]
risk_analysis = high_risk_projects.groupby(['project_type', 'region']).size().sort_values(ascending=False)
print("High-risk project combinations (Top 10):")
print(risk_analysis.head(10))

# Cost prediction for new projects
print(f"\nüîÆ COST PREDICTION FOR HYPOTHETICAL PROJECTS:")

# Create sample scenarios
scenarios = [
    {
        'name': 'Urban Housing Project',
        'initial_budget': 15.0,
        'project_type': 'Housing',
        'region': 'Urban',
        'total_area_sqft': 75000,
        'project_duration_months': 18
    },
    {
        'name': 'Rural Infrastructure Project',
        'initial_budget': 25.0,
        'project_type': 'Infrastructure',
        'region': 'Rural',
        'total_area_sqft': 120000,
        'project_duration_months': 36
    },
    {
        'name': 'Coastal Commercial Project',
        'initial_budget': 30.0,
        'project_type': 'Commercial',
        'region': 'Coastal',
        'total_area_sqft': 90000,
        'project_duration_months': 24
    }
]

print("Sample project cost predictions:")
for scenario in scenarios:
    # This is a simplified prediction - in practice, you'd need to engineer all features
    base_features = np.array([[
        scenario['initial_budget'],
        scenario['total_area_sqft'] / 10000,  # Normalized
        scenario['project_duration_months'],
        5.0,  # Average interest rate
        2.5,  # Average inflation
        75.0,  # Average quality score
        2.0,   # Average complexity
        1.5,   # Average risk
        100.0, # Economic index
        1.2    # Market favorability
    ]])
    
    # Note: This is a simplified example. In practice, you'd need to properly
    # encode categorical variables and include all engineered features
    print(f"  {scenario['name']}: Estimated cost range $XX.X - $XX.X million")
    print(f"    (Based on {scenario['project_type']} project in {scenario['region']} region)")

# ================================================================
# 11. ADVANCED RECOMMENDATIONS & ACTION ITEMS
# ================================================================

print(f"\nüìã STRATEGIC RECOMMENDATIONS & ACTION ITEMS")
print("=" * 50)

recommendations = [
    {
        'category': 'üí∞ Cost Optimization',
        'items': [
            'Focus on high-performing clusters with lower risk scores',
            'Implement stricter budget controls for project types with high overrun rates',
            'Prioritize projects in regions with better cost efficiency ratios',
            'Establish cost benchmarking systems based on project complexity scores'
        ]
    },
    {
        'category': 'üìä Risk Management',
        'items': [
            'Develop early warning systems using the risk score algorithm',
            'Implement enhanced monitoring for projects above 75th percentile risk threshold',
            'Create contingency reserves based on historical completion delay patterns',
            'Establish quality gates tied to the predictive quality score model'
        ]
    },
    {
        'category': 'üéØ Portfolio Optimization',
        'items': [
            'Balance portfolio across clusters to optimize risk-return profile',
            'Prioritize project types with highest ROI estimates in favorable regions',
            'Consider market timing based on interest rate and inflation predictions',
            'Implement dynamic resource allocation based on economic value index'
        ]
    },
    {
        'category': 'üìà Performance Monitoring',
        'items': [
            'Deploy real-time dashboards using the predictive models',
            'Establish KPIs based on feature importance rankings',
            'Create automated alerts for cost overrun probability',
            'Implement quarterly model retraining with new project data'
        ]
    }
]

for rec in recommendations:
    print(f"\n{rec['category']}:")
    for i, item in enumerate(rec['items'], 1):
        print(f"  {i}. {item}")

# ================================================================
# 12. MODEL DEPLOYMENT PREPARATION
# ================================================================

print(f"\nüöÄ MODEL DEPLOYMENT PREPARATION")
print("=" * 35)

# Save key model artifacts (simulation)
print("üíæ Saving model artifacts:")
print("  ‚úÖ Trained ensemble model")
print("  ‚úÖ Feature scaler and encoders")
print("  ‚úÖ Feature selection parameters")
print("  ‚úÖ Cluster model and parameters")
print("  ‚úÖ Model performance metrics")

# API endpoint structure
print(f"\nüåê Recommended API Endpoint Structure:")
api_structure = """
POST /api/v1/predict-project-cost
{
    "project_type": "Housing",
    "region": "Urban",
    "initial_budget": 15.0,
    "total_area_sqft": 75000,
    "project_duration_months": 18,
    "funding_source": "Federal",
    "local_gdp_growth": 2.8,
    "unemployment_rate": 5.2,
    "median_income": 65000,
    "interest_rate": 4.5,
    "inflation_rate": 2.5
}

Response:
{
    "predicted_cost": 18.7,
    "confidence_interval": [16.2, 21.3],
    "risk_score": 2.4,
    "cost_efficiency": 249.33,
    "cluster": 2,
    "recommendations": [...]
}
"""
print(api_structure)

# Monitoring recommendations
print(f"\nüìä Model Monitoring Recommendations:")
monitoring_items = [
    "Track prediction accuracy on new projects monthly",
    "Monitor feature drift in economic indicators",
    "Set up alerts for predictions outside historical ranges",
    "Retrain models quarterly with new project completion data",
    "A/B test model versions for continuous improvement"
]

for i, item in enumerate(monitoring_items, 1):
    print(f"  {i}. {item}")

print(f"\n" + "="*60)
print("üéâ ADVANCED FINANCIAL ANALYSIS COMPLETE!")
print(f"üìä Final Model Performance: R¬≤ = {r2_ensemble:.3f}")
print(f"üéØ Ready for production deployment and real-world application")
print("="*60)
```
